```python
from random import randint, uniform
```

```python
def generate_coords(n, minlat, maxlat, minlong, maxlong):
    coords = []
    for x in range(n):
        lat = uniform(minlat, maxlat)
        long = uniform(minlong, maxlong)
        coords.append([lat,long])
    return coords
```

```python
def plot(locations):
    
    avg1 = (max(list(map(lambda x: x[0], locations))) + min(list(map(lambda x: x[0], locations)))) / 2
    avg2 = (max(list(map(lambda x: x[1], locations))) + min(list(map(lambda x: x[1], locations)))) / 2
    
    folium_map = folium.Map(location=[avg1, avg2],
                        zoom_start=13,
                        tiles="Stamen Toner")
    
    for loc in locations:
        seed = randint(1, 10)
        if seed <= 3:
            color = "#E37222"
        else:
            color = "#0A8A9F"

        folium.CircleMarker(location = loc,
                                radius = 8,
                                color=color,
                                fill=True).add_to(folium_map)
    return folium_map  
```

```python
locations = generate_coords(100, 40.72, 40.76, -74, -73.98)
```

```python
plot(locations)
```
